```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
data=pd.read_csv("bottle.csv")
data
```

```python
data.drop(data.iloc[:, 0:5],inplace=True,axis=1)
data
```

```python
print(data.shape)
```

```python
data.drop(data.iloc[:,2:70],inplace=True,axis=1)
data
```

```python
mean_TdegC=data["T_degC"].mean()
data["T_degC"].fillna(mean_TdegC,inplace=True)
mean_salnity=data["Salnty"].mean()
data["Salnty"].fillna(mean_salnity,inplace=True)
data
```

```python
print(data.isnull())
```

```python
x=data[["Salnty"]]
y=data["T_degC"]
print(x)
print(x.shape)
#print(y)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
result=model.predict(x_test)
print(result)
```

```python
print("coeffictient",model.coef_)
print("intercept",model.intercept_)
print(data.corr())
```

```python
from sklearn.metrics import r2_score
r=r2_score(y_test,result)
print(r)
```
